```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Prostate_Cancer.csv")
```

```python
data.columns
```

```python
x = data[['radius', 'texture', 'perimeter', 'area',
       'smoothness', 'compactness', 'symmetry', 'fractal_dimension']].values
x
```

```python
y1 = data["diagnosis_result"]
y1
```

```python
y = []
for i in range(len(y1)):
    if y1[i] == "M":
        y.append([0])
    if y1[i] == "B":
        y.append([1])
y = np.array(y)
```

```python
knn = KNeighborsClassifier()
knn.fit(x,y)
```

```python
knn.predict([[2.300e+01, 1.200e+01, 1.510e+02, 9.540e+02, 1.430e-01, 2.780e-01,
        2.420e-01, 7.900e-02]])
```
